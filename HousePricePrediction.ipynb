```python
import pandas as pd
import pandas as pd
```

```python
data = pd.read_csv("Pune_house_data.csv")
```

```python
data.drop(columns = ['area_type','availability','society'],inplace = True)
```

```python
data.dropna(inplace =True)
```

```python
data
```

```python
data.site_location.value_counts()
```

```python
data["site_location"] = data["site_location"].apply(lambda x : x.strip())
```

```python
data.site_location.value_counts()
```

```python
location_stats = data.groupby("site_location")["site_location"].agg("count").sort_values(ascending =False)
```

```python
location_stats
```

```python
location_less_than_10_entries = location_stats[location_stats <= 10]
```

```python
location_less_than_10_entries
```

```python
data["site_location"]= data["site_location"].apply(lambda x : "other" if x in location_less_than_10_entries else x)
```

```python
data.site_location.value_counts()
```

```python
data 
```

```python
data['size'].unique()
```

```python
data['bedrooms']=data['size'].apply(lambda x : int (x.split(' ')[0]))
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    tokens = sqft.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    else:
        try:
            return float(sqft)
        except:
            return None

print(clean('200'))        # 200.0
print(clean('100-200'))    # 150.0
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.describe()
```

```python
data['sqft_per_bed'] = data['total_sqft']/data['bedrooms']
```

```python
data.sqft_per_bed.describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
```

```python
data2
```

```python
data2['price_per_sqft'] = data2['price']*100000/data2['total_sqft']
```

```python
data2
```

```python
data2.price_per_sqft.describe()
```

```python
data3 = data2[data2['price_per_sqft'] >= 2000]
```

```python
data3['price_per_sqft'].describe()
```

```python
data3
```

```python
data3.drop(columns = ['size','sqft_per_bed','price_per_sqft'],axis =1,inplace= True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['site_location']),remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
sr = StandardScaler()
```

```python
model = make_pipeline(col_trans,sr,lr)
```

```python
data_input = data3.drop(columns = ['price'])
data_output = data3['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_input,data_output,test_size = 0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([[1500.0,2.0,1.0,39.07,'Alandi Road',2]],columns=['total_sqft','bath','balcony','price','site_location','bedrooms'])
```

```python
model.predict(input)
```

```python
pk.dump(model,open('House_prediction_model.pkl','wb'))
```
